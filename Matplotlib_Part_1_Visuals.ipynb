# Type of Data <br>
**Numerical Data : Quantitative** <br>
**Categorical Data : Qualitative**



# Quantitative : Numerics : Measurable
<br> Expressed as Numbers
<br> Can be Counted
<br> Allows Mathematical Operations

Quantitative has TWO TYPES<br>
**1) Discrete :** single value, int, <br>
eg. number of toys,people,building,cars,items<br>
**2) Continouse :** continous value, may not be count on fingers real number within range,can be fractional,measures on instruments,<br>eg. Temprature, Scalse, Heights, Weights

**More types of Quantitative :**<br>

**1) Interval Data :** like interval from point to another point, No True Zero, eg. Tempratur from 35-45 but 0 temp does'nt emply that there is no temprature(absence of temp not proof),IQ,CGPA,Year,Month,Day-Time Interval No True Zero Proof<br>

**2) Ratio Data:** As name, you understand, can be Zero and absence of characteristic, ratio of Distance and Area,there is order and difference between values are meaningful,eg. Age, Weight,income,distance,counts,energy, speed, volume. All can be zero means they are absence

# Qualitative : Categorical :NonMeasurable<br>
Expressed as Charasteristics<br>
Can be understand not count<br>
Don't Allows Mathematical Operation either in number form<br>

Qualitative has TWO TYPES<br>
**1) Nominal :** category , distinct and seperate, independant, order not matters<br>eg. colors, types,Gender, we can thing any numericals from red,male, <br>
[data category that not belongs to thought of numbers]
<br>
**2) Ordinal :** category, can be same, statisfaction stars, educational level, rankings all can be same for records.
[data category that we can think who is first who is big who is small]

Remeber: If you write code on single cells, so the graph overlaped on each other.

**In Matplolib:** There is no function for multi bar graph, so we use some tricks

```python
#Lets take array
plt.bar(np.arange(df.shape[0])-0.25,df['2015'],width=0.25,color='red') #Then copy paste
plt.bar(np.arange(df.shape[0]),df['2016'],width=0.25,color='green')
plt.bar(np.arange(df.shape[0])+0.25,df['2017'],width=0.25,color='blue')

plt.xticks(np.arange(df.shape[0]),df['batsman']) #From this it take existing label than write what column you want to replace
plt.show()
```

```python
#Take Another Dataset of Batsman
df=pd.read_csv('/content/gayle-175.csv')
df
```

```python
plt.pie(df['batsman_runs'])
```

```python
#add Labels
plt.pie(df['batsman_runs'],labels=df['batsman'])
```

```python
#Add Data Labels autopct
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%')
plt.show()
```

```python
#Add colors Manually
color=['red','blue','green','yellow','gray','pink']
plt.pie(df['batsman_runs'],labels=df['batsman'],colors=color)
```

```python
#We can Alos Take any Particular Part from PIE
#explode parameter can take any float small value 0.1 to 1 uses is recommended

plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%',explode=[1,0,0,0,0,0])
plt.show()
```

```python
#Shadow Parameter
#Shadow=True
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%',explode=[1,0,0,0,0,0],shadow=True)
plt.show()
```
